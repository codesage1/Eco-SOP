```python

df2 = pd.read_excel(file_path, sheet_name="CP")  # Sheet numbers are 0-indexed

# Create the 'has_null' column
#df2['has_null'] = df2.isnull().any(axis=1)
#df2.loc[df.any(axis=1), 'has_null'] = False  # If any value in the row is True, set to False
df2.head()
# Save the modified DataFrame
#df2.to_excel("modified_file.xlsx", index=False)
```

```python

# Find common companies
common_companies = set(df['Company']).intersection(set(df2['Company']))

# Filter both DataFrames to keep only common companies
df1_filtered = df[df['Company'].isin(common_companies)]
df2_filtered = df2[df2['Company'].isin(common_companies)]
```
